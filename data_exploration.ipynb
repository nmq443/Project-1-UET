# Data exploration
This notebook explores the dataset used for the project. Link to the dataset: https://universe.roboflow.com/vehicle-mscoco/vehicles-coco/dataset/2

## Import libraries

```python
import numpy as pd
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pylabel import importer
```

```python
# Load dataset (in YOLO format)
training_data = importer.ImportYoloV5(
    path='./vehicles-coco-dataset/train/labels/',
    img_ext='jpg',
    cat_names=['bus', 'car', 'motorcycle', 'truck'],
    path_to_images='../../train/images/'
)

valid_data = importer.ImportYoloV5(
    path='./vehicles-coco-dataset/valid/labels/',
    img_ext='jpg',
    cat_names=['bus', 'car', 'motorcycle', 'truck'],
    path_to_images='../../valid/images/'
)

test_data = importer.ImportYoloV5(
    path='./vehicles-coco-dataset/test/labels/',
    img_ext='jpg',
    cat_names=['bus', 'car', 'motorcycle', 'truck'],
    path_to_images='../../test/images/'
)
```

```python
# Get dataframe
training_df = training_data.df
valid_df = valid_data.df
test_df = test_data.df

# See 5 samples of training data
training_df.sample(5)
```

```python
dataset = pd.concat([training_df, valid_df, test_df])
dataset.sample(5)
```

```python
dataset.columns
```

```python
dataset = dataset.dropna(axis=1)
dataset.sample(5)
```

```python
dataset = dataset.drop(columns=['ann_segmentation', 'ann_iscrowd',
       'ann_keypoints', 'ann_pose', 'ann_truncated', 'ann_difficult', 'cat_supercategory', 'split'])
```

```python
dataset.sample(5)
```

## The collumns corresponds to

* img_folder: the path of the corresponding image (with ../../ before because of an issue in pylabel)
* img_filename: the name of the corresponding image
* img_path: not used here, but this field can be useful for some format of dataset. For the VOC format, this field is not used.
* img_id: an id given by pylabel to the image
* img_width: the width of the images
* img_height: the height of the images
* img_depth: the color channel of the images
* ann_bbox_xmin: the minimum x coordinate of a bounding boxe in the image
* ann_bbox_ymin: the minimum y coordinate of a bounding boxe in the image
* ann_bbox_xmax: the maximum x coordinate of a bounding boxe in the image
* ann_bbox_ymax: the maximum y coordinate of a bounding boxe in the image
* ann_bbox_width: the bounding box width
* ann_bbox_height: the bounding box height
* ann_area: the bounding boxe area
* cat_id: contains the category ID (here only 0 as we only have one category, one class)
* cat_name: the name of the class 
* annotated: 1 if the image is annoted, 0 if not

Note: This dataframe will contains many duplicates of the images because one row corresponds to one bounding box (one label). So if an image contains 3 bounding boxes , then 3 rows will have the same filename because it's the same image

Distribution of the positions of the bounding boxes

```python
fig, axes = plt.subplots(1, 2, figsize=(15, 5))

sns.histplot(data=dataset.ann_bbox_width, ax=axes.flatten()[0])
sns.histplot(data=dataset.ann_bbox_height, ax=axes.flatten()[1])

bbox_width_dist = sns.histplot(data=dataset.ann_bbox_width, ax=axes.flatten()[0]).get_figure()
bbox_height_dist = sns.histplot(data=dataset.ann_bbox_height, ax=axes.flatten()[1]).get_figure()

bbox_width_dist.savefig('./figures/bbox_width_distribution.png')
bbox_height_dist.savefig('./figures/bbox_height_distribution.png')
```

Most of the bounding boxes have low width (between 0 and 100 pixel). The height is similar. Thanks to the .describe() did previously, we can see that 76% of the bounding boxes have a height less than 97 and width less than 96.5. This is very informative because it shows that a large percentage of the bounding boxes will be small-moderate. It means that a lot of images will contains vehicles that are far away from the camera, or at least not close by. Thus, we can suppose that our model will be better at recognizing vehicles that are not too close to the camera. After validating the results and testing the model in a real environement, we will probably be able to answer this question. In that case, the camera will have to be at a certain range from the road in order to detect the vehicles with a good accuracy.
