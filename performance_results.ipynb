# This notebook is used for visualizing the performance of the model

## Import libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

## Get the result using pandas.Dataframe

```python
result = pd.read_csv('./results/runs/detect/train/results.csv')
```

```python
result.head()
```

```python
result.columns
```

```python
train_box_loss = result['train/box_loss']
train_cls_loss = result['train/cls_loss']
train_dfl_loss = result['train/dfl_loss']

precision = result['metrics/precision(B)']
recall = result['metrics/recall(B)']
mAP50 = result['metrics/mAP50(B)']
mAP50_95 = result['metrics/mAP50-95(B)']

val_box_loss = result['val/box_loss']
val_cls_loss = result['val/cls_loss']
val_dfl_loss = result['val/dfl_loss']
```

## Visualize the training loss (consists of box loss, classes loss and distribution focal loss)

```python
train_fig = plt.figure(figsize=(15, 5))
plt.subplot(131)
plt.plot(train_box_loss)
plt.xlabel('Epochs')
plt.ylabel("Training's box loss")

plt.subplot(132)
plt.plot(train_cls_loss)
plt.xlabel('Epochs')
plt.ylabel("Training's class loss")

plt.subplot(133)
plt.plot(train_dfl_loss)
plt.xlabel('Epochs')
plt.ylabel("Training's distribution focal loss")

train_fig.savefig('./figures/performances/train_loss.png')
```

## Visualize the validation loss (consists of box loss, classes loss and distribution focal loss)

```python
plt.figure(figsize=(15, 5))
plt.subplot(131)
plt.plot(val_box_loss)
plt.xlabel('Epochs')
plt.ylabel("Validation's box loss")

plt.subplot(132)
plt.plot(val_cls_loss)
plt.xlabel('Epochs')
plt.ylabel("Validation's class loss")

plt.subplot(133)
plt.plot(val_dfl_loss)
plt.xlabel('Epochs')
plt.ylabel("Validation's distribution focal loss")

train_fig.savefig('./figures/performances/val_loss.png')
```

## Visualize the mAP, precision and recall

```python
precision = result['metrics/precision(B)']
recall = result['metrics/recall(B)']
mAP50 = result['metrics/mAP50(B)']
mAP50_95 = result['metrics/mAP50-95(B)']

plt.figure(figsize=(15, 5))
plt.subplot(121)
plt.plot(precision)
plt.xlabel('Epochs')
plt.ylabel("Precision")

plt.subplot(122)
plt.plot(recall)
plt.xlabel('Epochs')
plt.ylabel("Recall")
train_fig.savefig('./figures/performances/precision-recall.png')

plt.figure(figsize=(15, 5))
plt.subplot(121)
plt.plot(mAP50)
plt.xlabel('Epochs')
plt.ylabel("Mean average precision (50)")

plt.subplot(122)
plt.plot(mAP50_95)
plt.xlabel('Epochs')
plt.ylabel("Mean average precision (95)")
train_fig.savefig('./figures/performances/metrics-mAP.png')
```
